```python
#import packages

import pandas as pd
import numpy as np
```

```python
#read in data

hub_nash = pd.read_csv('../data.nosync/hubNashville__311__Service_Requests.csv');
hub_nash.head()
```

```python
hub_nash.columns
```

```python
hub_nash = hub_nash.drop(columns = ['System of Origin ID',
       'Contact Type', 'Preferred Language', 'State Issue', 'Parent Request'])
```

```python
hub_nash.columns
```

```python
hub_nash.columns = ['request#', 'status', 'request_type', 'subrequest_type',
       'add_subrequest_type', 'opened',
       'closed', 'origin', 'closed_when_created',
       'address', 'city', 'council_district', 'zip', 'lat', 'lng',
       'mapped']
hub_nash.head()
```

```python
t1 = pd.to_datetime(hub_nash['closed'])
t2 = pd.to_datetime(hub_nash['opened'])
hub_nash['time_taken'] = t1-t2
```

```python
col_name = 'time_taken'
hub_time = hub_nash.pop(col_name)
hub_nash.insert(7, col_name, hub_time)
```

```python
hub_nash.head()
```

```python
# insert the datetime variables into the df

hub_nash = hub_nash.drop(columns = ['opened', 'closed'])
hub_nash.insert(5, 'opened', t2, allow_duplicates = False)
hub_nash.insert(6, 'closed', t1, allow_duplicates = False)
```

```python
hub_nash.dtypes
```

```python
hub_nash['season'] = pd.cut(
    (hub_nash.opened.dt.dayofyear + 11) % 366,
    [0, 91, 183, 275, 366],
    labels=['winter', 'spring', 'summer', 'fall']
)
```

```python
hub_nash.head()
```

```python
hub_nash['month'] = hub_nash['opened'].dt.month
hub_nash.head(1)
```

```python
dayofweek={0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
hub_nash['day'] = hub_nash['opened'].dt.weekday.map(dayofweek)
hub_nash.head()
```

```python
b = [0,4,11,16,20,24]
l = ['late_night', 'morning','afternoon','evening','night']
hub_nash['time_of_day'] = pd.cut(hub_nash['opened'].dt.hour, bins=b, labels=l)
hub_nash.head(1)
```
